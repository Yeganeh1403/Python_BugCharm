### Question 04
Create a 30×5 matrix (30 students, 5 courses).

* Compute each student's average.
* Compute each course average.
* Identify students who scored above 15 in at least 3 courses.
* Find the hardest course (with the loest scores)

```python
array_04 = np.random.randint(0, 20, size=(30, 5))
print(np.round(np.mean(array_04, axis=1), 1))
print(np.round(np.mean(array_04, axis=0), 2))
more_than_15 = np.where(array_04 >= 15, 1, 0)
print(np.where(np.sum(more_than_15, axis=1) > 2))
print(np.where(np.mean(array_04, axis=0) == np.min(np.mean(array_04, axis=0))))
```

### Question 06

Create a Series with 30 different product names as indexes

and enter random numbers (sales) as values.

* Find best-selling product.
* Compute percentage share of each product.
* Remove products below average sales.

```python
sales_series = pd.Series(np.random.randint(100, 1001, 30),
index=[f"product{i}" for i in range(1, 31)])
min_sale = min(sales_series)
print("product with the least price is:\n", sales_series[sales_series == min_sale].index)
#we concider 20% share for each product
print("percentage share of each product:\n", sales_series * (20 / 100))
mean_sale = sales_series.mean()
print("mean price is:\n", mean_sale)
sales_series = sales_series[sales_series > mean_sale]
print("products with price more than average sales:\n", sales_series)
```

### Question 08

Create a Series of 50 people birthday year (1925 to 2025)

and set indexes with their gender (random Female/Male).

then, separate these 50 pople into 6 different series based on their generation:

* Alpha: Born 2013–2025.
* Z: Born 1997–2012 (Ages 14–29).
* Millennials: Born 1981–1996 (Ages 30–45).
* X: Born 1965–1980 (Ages 46–61).
* Baby Boomers: Born 1946–1964 (Ages 62–80).
* Silent: Born 1925–1945 (Ages 81–101). 

Finally, print number of Female and Male popele in each generation.

```python
birth_year_series = pd.Series(np.random.randint(1925, 2026, 50), index=[np.random.choice(["male", "female"], 50)])
```

```python
Alpha = {"male": 0, "female": 0}
Z = {"male": 0, "female": 0}
Millennials = {"male": 0, "female": 0}
X = {"male": 0, "female": 0}
BabyBoomers = {"male": 0, "female": 0}
Silent = {"male": 0, "female": 0}
```

```python
for i, j in birth_year_series.items():
    if j >= 2013 and j <= 2025:
        if i == "male":
            Alpha["male"] += 1
        if i == "female":
            Alpha["female"] += 1
    elif j >= 1997 and j <= 2012:
        if i == "male":
            Z["male"] += 1
        if i == "female":
            Z["female"] += 1
    elif j >= 1981 and j <= 1996:
        if i == "male":
            Millennials["male"] += 1
        if i == "female":
            Millennials["female"] += 1
    elif j >= 1965 and j <= 1980:
        if i == "male":
            X["male"] += 1
        if i == "female":
            X["female"] += 1
    elif j >= 1946 and j <= 1964:
        if i == "male":
            BabyBoomers["male"] += 1
        if i == "female":
            BabyBoomers["female"] += 1
    elif j >= 1925 and j <= 1945:
        if i == "male":
            Silent["male"] += 1
        if i == "female":
            Silent["female"] += 1
```
